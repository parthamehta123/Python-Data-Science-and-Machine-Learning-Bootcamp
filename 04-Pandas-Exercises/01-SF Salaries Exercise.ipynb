```python
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
from sklearn import tree
import matplotlib.pyplot as plt

df = pd.read_csv('/Users/parthamehta/python-workspace/Data.csv')

print(df.head(5))
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, 5]

print('Data')
print(X.head())
print('Target')
print(y.head())

X = pd.get_dummies(X)
print(X.columns)

features = ['Age', 'Income_High', 'Income_Low', 'Income_Medium', 'Student_No',
'Student_Yes', 'Credit_Rating_Excellent', 'Credit_Rating_Fair']


clf = DecisionTreeClassifier(random_state=0)

clf.fit(X, y)

fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (6,6), dpi=1200)

tree.plot_tree(clf,
feature_names=features,
filled=True)

fig.savefig('/Users/parthamehta/python-workspace/res.png')
```

** How many people have the word Chief in their job title? (This is pretty tricky) **

```python
def chief_string(title):
    if 'chief' in title.lower():
        return True
    else:
        return False
```

```python
sum(sal['JobTitle'].apply(lambda x: chief_string(x)))
```
